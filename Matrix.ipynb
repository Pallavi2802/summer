```python
# Matrix multiplication
m = int(input('number of rows, m = '))
n = int(input('number of columns, n = '))
matrix1=[[0 for var in range  (0 , n)] for i in range (0 , m)]
matrix2=[[0 for var in range  (0 , n)] for i in range (0 , m)]
result=[[0 for var in range  (0 , n)] for i in range (0 , m)]
print("enteries in matrix 1 ")
for i in range (0,m):
    for j in range (0,n):
        print ('entry in row: ',i+1,' column: ',j+1)
        matrix1[i][j] = int(input())
print (matrix1)

print("enteries in matrix 2 ")

for i in range (0,m):
    for j in range (0,n):
        print ('entry in row: ',i+1,' column: ',j+1)
        matrix2[i][j] = int(input())
print (matrix2)

for i in range (0,m):
    for j in range (0,n):
        for k in range(0,n) :
            result[i][j] += matrix1[i][k] * matrix2[k][j]
print("Multiplication of two matrix is ")
print (result)
```

```python
# matrix Addition
m = int(input('number of rows, m = '))
n = int(input('number of columns, n = '))
matrix1=[[0 for var in range  (0 , n)] for i in range (0 , m)]
matrix2=[[0 for var in range  (0 , n)] for i in range (0 , m)]
result=[[0 for var in range  (0 , n)] for i in range (0 , m)]
print("enteries in matrix 1 ")
for i in range (0,m):
    for j in range (0,n):
        print ('entry in row: ',i+1,' column: ',j+1)
        matrix1[i][j] = int(input())
print (matrix1)

print("enteries in matrix 2 ")

for i in range (0,m):
    for j in range (0,n):
        print ('entry in row: ',i+1,' column: ',j+1)
        matrix2[i][j] = int(input())
print (matrix2)

for i in range (0,m):
    for j in range (0,n):
        result[i][j] += matrix1[i][j] + matrix2[i][j]
print("Addition of two matrix is ")
print (result)
```
