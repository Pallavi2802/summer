```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

food = ctrl.Antecedent(np.arange(0, 11, 1), 'food')
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')
```

```python
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
```

```python
food['bad']=fuzz.trimf(food.universe,[0,0,5])
food['medium']=fuzz.trimf(food.universe,[0,5,10])
food['excellent']=fuzz.trimf(food.universe,[5,10,10])
```

```python
service['poor']=fuzz.trimf(food.universe,[0,0,5])
service['average']=fuzz.trimf(food.universe,[0,5,10])
service['excellent']=fuzz.trimf(food.universe,[5,10,10])
```

```python
food.view()
plt.title("Food Qulaity")
plt.show()
```

```python
service.view()
plt.title("service Qulaity")
plt.show()
```

```python
tip.view()
plt.title("tip amount")
plt.show()
```

```python
rule1=ctrl.Rule(antecedent=food['bad'] & service['poor'],consequent=tip['low'])
```

```python
rule2=ctrl.Rule(antecedent=service['average'],consequent=tip['medium'])
```

```python
rule3=ctrl.Rule(antecedent=food['excellent'] | service['excellent'],consequent=tip['high'])
```

```python
rules=[rule1,rule2,rule3]
```

```python
rule1.view()
plt.title("rules")
plt.show()
```

```python
tipping_ctrl=ctrl.ControlSystem(rules)
```

```python
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
```

```python
inputdic={'food':6.5,'service':9.8}
```

```python
tipping.inputs(inputdic)
```

```python
tipping.compute()
```

```python
tip.view(sim=tipping)
```

```python
print(tipping.output['tip'])
```

```python
tipamt=tipping.output['tip']
```

```python
tipamt=float(format(tipamt,'.2f'))
```

```python
print(tipamt)
```

```python
tip.view(sim=tipping)
plt.title("tip to be paid")
plt.text(10,0.3,"tipAmt:"+ str(tipamt))
plt.show()
```
